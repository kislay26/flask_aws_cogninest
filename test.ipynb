```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
DATA_DIR = "data"
CLEANED_CSV = os.path.join(DATA_DIR, "imdb_reviews_cleaned.csv")
```

```python
df = pd.read_csv(CLEANED_CSV)
df.shape
```

```python
def perform_eda(df):
    # Sentiment Distribution
    sentiment_counts = df['sentiment'].value_counts()
    print("Sentiment Distribution:")
    print(sentiment_counts)
    
    # Average Review Length
    df['review_length'] = df['review_text'].apply(lambda x: len(x.split()))
    avg_length = df.groupby('sentiment')['review_length'].mean()
    print("\nAverage Review Length (in words):")
    print(avg_length)
    
    # Plotting Sentiment Distribution
    sentiment_counts.plot(kind='bar', title='Sentiment Distribution')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.show()
    
    # Common Words in Positive Reviews
    positive_reviews = df[df['sentiment'] == 'positive']['review_text']
    words = ' '.join(positive_reviews).split()
    common_words = Counter(words).most_common(10)
    print("\nTop 10 Common Words in Positive Reviews:")
    print(common_words)
```

```python
perform_eda(df)
```
